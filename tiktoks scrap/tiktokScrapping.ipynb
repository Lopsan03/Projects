```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

url = 'https://www.tiktok.com/@soneedo'

# Initialize the Selenium web driver
driver = webdriver.Chrome()

# Open the URL
driver.get(url)

# Wait for the page to load (you may need to adjust the waiting time)
time.sleep(20)

# Function to scroll down the page
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)  # Adjust the sleep time as needed

# Scroll down the page multiple times to load more videos
max_scrolls = 15  # Adjust the maximum number of scrolls as needed
for _ in range(max_scrolls):
    scroll_down(driver)

# Find video elements with the updated CSS selector
videos = driver.find_elements(By.CSS_SELECTOR, 'a.css-1wrhn5c-AMetaCaptionLine.eih2qak0')

# Find video view count elements with the updated CSS selector
videos_views = driver.find_elements(By.CLASS_NAME, 'video-count.css-dirst9-StrongVideoCount.e148ts222')

# Arrays for the video links and views
links = []
views = []

# Extract hrefs and video views
for video, video_view in zip(videos, videos_views):
    video_link = video.get_attribute('href')
    view = video_view.text
    if view:
        if 'M' in view:
            view = int(float(view.replace('M', '')) * 1_000_000)
        elif 'K' in view:
            view = int(float(view.replace('K', '')) * 1_000)
        else:
            view = int(view.replace(',', ''))
        links.append(video_link)
        views.append(view)

# Close the web driver
driver.quit()

# Print the extracted links and views
for link, view in zip(links, views):
    print(f"Video Link: {link}\nViews: {view}\n")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from datetime import datetime
import time
import json

# Initialize the Selenium web driver
driver = webdriver.Chrome()
driver.maximize_window()

# List to store extracted data
video_data = []

for link in links:
    # Open the URL
    driver.get(link)

    # Wait for the page to load (you may need to adjust the waiting time)
    time.sleep(5)

    # Find video elements
    video_duration_element = driver.find_element(By.CSS_SELECTOR, 'div.css-1cuqcrm-DivSeekBarTimeContainer.e1ya9dnw1')
    video_duration_text = video_duration_element.text
    video_duration = video_duration_text.split(" / ")[1]  # Extract the duration from the text
    
    
    # Find the likes element
    likes_element = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e="like-count"].css-n6wn07-StrongText.edu4zum2')
    video_likes = likes_element.text
    
    # Find the comments element
    #<strong data-e2e="comment-count" class="css-n6wn07-StrongText edu4zum2">2</strong>
    comments_element = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e="comment-count"].css-n6wn07-StrongText.edu4zum2')
    video_comments = comments_element.text
    
    saved_element = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e="undefined-count"].css-n6wn07-StrongText.edu4zum2')
    video_saved = saved_element.text
    
    # Function to scroll down a specific percentage of the page
    def scroll_down(driver, percentage=40):
        current_height = driver.execute_script("return window.scrollY")
        window_height = driver.execute_script("return window.innerHeight")
        target_height = current_height + (window_height * percentage / 100)
        driver.execute_script(f"window.scrollTo(0, {target_height});")
        time.sleep(5)  # Adjust the sleep time as needed

    # Scroll down 20% of the page to load more content
    scroll_down(driver, percentage=40)

    # Find the music element
    music_element = driver.find_element(By.CSS_SELECTOR, 'div.css-pvx3oa-DivMusicText.epjbyn3')
    video_music = music_element.text

    # Extract video description
    video_description_element = driver.find_element(By.CSS_SELECTOR, 'h1[data-e2e="browse-video-desc"]')
    
    # Extract text from the description element
    video_description = video_description_element.text

    # Extract hashtags from video description
    video_hashtags = [elem.strip() for elem in video_description.split() if elem.startswith("#")]

    # Find the script element
    script_element = driver.find_element(By.ID, '__UNIVERSAL_DATA_FOR_REHYDRATION__')

    # Extract the content inside the script element
    script_content = script_element.get_attribute("innerHTML")
    # Convert the JSON string to a JSON object
    data = json.loads(script_content)

    create_time_str = data.get("__DEFAULT_SCOPE__", {}).get("webapp.video-detail", {}).get("itemInfo", {}).get("itemStruct", {}).get("createTime")

    # Convert string to integer
    create_time = int(create_time_str)

    # Convert Epoch to Date-Time
    datetime_obj = datetime.fromtimestamp(create_time)

    # Append data to the list
    video_data.append({
        'Link': link,
        'Duration': video_duration,
        'Music': video_music,
        'Likes': video_likes,
        'Comments': video_comments,
        'Saved': video_saved,
        'Hashtags': video_hashtags,
        'DatePosted': datetime_obj
    })

# Close the web driver
driver.quit()

# Print the extracted data
for data in video_data:
    print(f'Link: {data["Link"]}')
    print(f'Duration: {data["Duration"]}')
    print(f'Music: {data["Music"]}')
    print(f'Likes: {data["Likes"]}')
    print(f'Comments: {data["Comments"]}')
    print(f'Saved: {data["Saved"]}')
    print(f'Hashtags: {data["Hashtags"]}')
    print(f'Date Posted: {data["DatePosted"]}\n')





```

```python
import pandas as pd


# Custom function to convert 'K' and 'M' to integers
def convert_to_int(value_str):
    if 'K' in value_str:
        return int(float(value_str.replace('K', '')) * 1000)
    elif 'M' in value_str:
        return int(float(value_str.replace('M', '')) * 1000000)
    else:
        return int(value_str)

# Apply the custom function to the 'Views' and 'Likes' columns
df['Likes'] = df['Likes'].apply(convert_to_int)
df['Comments'] = df['Comments'].apply(convert_to_int)
df['Saved'] = df['Saved'].apply(convert_to_int)

print(df)
```

```python
# Calculate the 'LikesToViewsRatio' column
df['LikesToViewsRatio'] = (df['Likes'].astype(int) / df['views']).round(2)

# Calculate the correlation between 'Likes' and 'views'
correlation_likes_views = df['Likes'].astype(int).corr(df['views'].astype(int))

# Calculate the correlation between 'Comments' and 'views'
correlation_comments_views = df['Comments'].astype(int).corr(df['views'].astype(int))

# Calculate the correlation between 'LikesToViewsRatio' and 'views'
correlation_likes_to_views = df['LikesToViewsRatio'].corr(df['views'].astype(int))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract the day of the week
df['DayOfWeek'] = df['DatePosted'].dt.day_name()

# Calculate total views for each day of the week
views_by_day = df.groupby('DayOfWeek')['views'].sum()

# Sort the days of the week in the correct order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
views_by_day = views_by_day.reindex(days_order)

# Create a bar chart to visualize views by day of the week
plt.figure(figsize=(10, 6))
views_by_day.plot(kind='bar', color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Total Views')
plt.title('Total Views by Day of the Week')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract the hour and minute
df['HourOfDay'] = df['DatePosted'].dt.hour
df['Minute'] = df['DatePosted'].dt.minute

# Calculate the time in half-hour intervals
df['HalfHour'] = df['HourOfDay'] + df['Minute'] / 60

# Calculate total views for each half-hour interval
views_by_half_hour = df.groupby('HalfHour')['views'].sum()

# Create a line chart to visualize views by half-hour intervals
plt.figure(figsize=(10, 6))
views_by_half_hour.plot(kind='line', marker='o', color='skyblue')
plt.xlabel('Time (Half-Hour Intervals)')
plt.ylabel('Total Views')
plt.title('Total Views by Half-Hour Intervals')
plt.xticks(range(int(df['HalfHour'].max()) + 1))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Convert the 'Duration' column to strings
df['Duration'] = df['Duration'].astype(str)

# Initialize columns for minutes and seconds
df['Minutes'] = 0
df['Seconds'] = 0

# Split the 'Duration' column into minutes and seconds (if available)
for index, row in df.iterrows():
    duration_parts = row['Duration'].split(':')
    try:
        minutes, seconds = map(int, duration_parts)
    except ValueError:
        minutes, seconds = 0, int(duration_parts[0])
    df.at[index, 'Minutes'] = minutes
    df.at[index, 'Seconds'] = seconds

# Calculate the total duration in seconds
df['Duration'] = (df['Minutes'] * 60) + df['Seconds']

# Create a histogram to visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Duration'], bins=15, density=True, alpha=0.6, color='b', label='Video Duration Distribution')

# Calculate mean and standard deviation for a normal distribution curve
mu, std = df['Duration'].mean(), df['Duration'].std()

# Create a normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution Curve')
plt.title('Video Duration Distribution and Normal Distribution Curve')
plt.xlabel('Duration (s)')
plt.ylabel('Frequency')

# Find the best duration (mean of the normal distribution curve)
best_duration = mu
plt.axvline(best_duration, color='r', linestyle='--', label='Best Duration (Mean)')

plt.legend()
plt.show()

print(f"Best Video Duration (in seconds) for Highest Average Views: {best_duration:.2f} seconds")
```
