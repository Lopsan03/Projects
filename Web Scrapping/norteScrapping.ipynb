# Imports

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import pandas as pd
import time
import re
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.support.ui import Select
```

# Setting up webdriver and initial link

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def set_filters(driver, inmueble=None, estado=None, zona=None, colonia=None,
                precio_inicial=None, precio_final=None, m2_terreno_inicial=None,
                m2_terreno_final=None, m2_construccion_inicial=None, 
                m2_construccion_final=None, plantas=None, banios=None, 
                recamaras=None, solo_foto=False):
    
    # Set Inmueble
    if inmueble:
        select_inmueble = Select(driver.find_element(By.ID, 'idinmueble'))
        select_inmueble.select_by_value(inmueble)
    
    # Set Estado
    if estado:
        select_estado = Select(driver.find_element(By.ID, 'Estado'))
        select_estado.select_by_value(estado)
    
    # Set Zona
    if zona:
        select_zona = Select(driver.find_element(By.ID, 'Zonas'))
        select_zona.select_by_value(zona)
    
    # Set Colonia
    if colonia:
        select_colonia = Select(driver.find_element(By.ID, 'Colonia'))
        select_colonia.select_by_value(colonia)
    
    # Set Precio Inicial
    if precio_inicial:
        select_precio_inicial = Select(driver.find_element(By.ID, 'PrecioInicial'))
        select_precio_inicial.select_by_value(precio_inicial)
    
    # Set Precio Final
    if precio_final:
        select_precio_final = Select(driver.find_element(By.ID, 'PrecioFinal'))
        select_precio_final.select_by_value(precio_final)
    
    # Set m² Terreno Inicial
    if m2_terreno_inicial:
        select_terreno_inicial = Select(driver.find_element(By.ID, 'TerrenoInicial'))
        select_terreno_inicial.select_by_value(m2_terreno_inicial)
    
    # Set m² Terreno Final
    if m2_terreno_final:
        select_terreno_final = Select(driver.find_element(By.ID, 'TerrenoFinal'))
        select_terreno_final.select_by_value(m2_terreno_final)
    
    # Set m² Construcción Inicial
    if m2_construccion_inicial:
        select_construccion_inicial = Select(driver.find_element(By.ID, 'ConstruccionInicial'))
        select_construccion_inicial.select_by_value(m2_construccion_inicial)
    
    # Set m² Construcción Final
    if m2_construccion_final:
        select_construccion_final = Select(driver.find_element(By.ID, 'ConstruccionFinal'))
        select_construccion_final.select_by_value(m2_construccion_final)
    
    # Set Plantas
    if plantas:
        select_plantas = Select(driver.find_element(By.ID, 'Plantas'))
        select_plantas.select_by_value(plantas)
    
    # Set Baños
    if banios:
        select_banios = Select(driver.find_element(By.ID, 'Banios'))
        select_banios.select_by_value(banios)
    
    # Set Recámaras
    if recamaras:
        select_recamaras = Select(driver.find_element(By.ID, 'Recamaras'))
        select_recamaras.select_by_value(recamaras)
    
    # Set 'Solo Avisos con Fotografía'
    if solo_foto:
        checkbox_foto = driver.find_element(By.ID, 'ChbFoto')
        if not checkbox_foto.is_selected():
            checkbox_foto.click()
    
    # Click the search button
    search_button = driver.find_element(By.CSS_SELECTOR, '.btnSearch a')
    search_button.click()

# Example usage with WebDriver
driver = webdriver.Chrome()

# Open the website
driver.get("https://www.avisosdeocasion.com/venta-casas.aspx?PlazaBusqueda=2&Plaza=2&utm_campaign=include_avisos_en_portada_elnorte&utm_source=categorias&utm_medium=bienes_raices")

# Apply filters and get the final URL
set_filters(driver, inmueble="3", estado="2", zona="17", colonia="0", solo_foto=True)

# Extract the final page URL and save it in a variable named URL
URL = driver.current_url
```

# Extract Links of houses

```python
# Extract all the links to the individual house pages
house_links = []

# Get the total number of pages from the pagination
pagination_element = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_divcpx2")
total_pages = len(pagination_element.find_elements(By.CLASS_NAME, "cpa")) + 1  # Includes the current page (e.g., 1, 2, 3...)

# Loop through each page and collect the links
for page_num in range(1, total_pages + 1):
    print(f"Processing page {page_num}...")
    driver.get(f"{URL}&Pagina={page_num}")
    time.sleep(2)  # Add a short delay to allow the page to load

    # Extract links on the current page
    elements = driver.find_elements(By.CSS_SELECTOR, 'a.tituloresult')
    for elem in elements:
        house_links.append(elem.get_attribute('href'))

# Ensure house_links is unique
house_links = list(set(house_links))

# Reduce default timeout to 3 seconds
timeout = 3
```

# Date setting function

```python
# Function to convert the publication date string to a datetime object
def convert_to_date(date_str):
    month_map = {
        'Ene': '01',
        'Feb': '02',
        'Mar': '03',
        'Abr': '04',
        'May': '05',
        'Jun': '06',
        'Jul': '07',
        'Ago': '08',
        'Sep': '09',
        'Oct': '10',
        'Nov': '11',
        'Dic': '12'
    }
    date_str = date_str.replace('Publicado el ', '')
    day, month = date_str.split(' de ')
    date_str = f"{datetime.now().year}-{month_map[month]}-{int(day):02d}"
    return datetime.strptime(date_str, '%Y-%m-%d')
```

# Loop through links and extract data

```python
# Create a list to store the data for each house
houses_data = []

# Loop through each house link to extract the relevant details
for idx, link in enumerate(house_links):
    print(f"Processing link {idx + 1}/{len(house_links)}: {link}")

    driver.get(link)

    # Check if there's a webpage link (indicating it's promoted by a real estate company)
    try:
        if WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.XPATH, "//td[@class='carac_td']/a[contains(@href, 'http')]"))
        ):
            print(f"Skipping link {idx + 1}/{len(house_links)}: Real estate company found")
            continue
    except TimeoutException:
        pass

    try:
        info_paragraph = WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.XPATH, "//div[@id='infocompleta']"))
        ).text.strip()

        # Check for 'RAICES', 'inmuebles', keywords and non-gmail/hotmail emails
        if any(keyword in info_paragraph.upper() for keyword in ['RAICES', 'INMUEBLES', 'REAL ESTATE', 'ASESORA', 'ASESOR','81-1865-1381', '81-1311-9696', '81-1010-9417', '811-817-19-83', '818-064-51-65', '818-252-79-23', '81-2317-6061', '81-1068-14-51']):
            print(f"Skipping link {idx + 1}/{len(house_links)}: Real estate keywords found")
            continue

        emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', info_paragraph)
        if any(not email.endswith(('@gmail.com', '@hotmail.com')) for email in emails):
            print(f"Skipping link {idx + 1}/{len(house_links)}: Non-gmail/hotmail email found")
            continue
    except TimeoutException:
        pass

    # Extract house details
    house_details = {}
    xpaths = {
        'Zone': "//td[@style='width:65%;']/h2[contains(text(), 'ZONA')]/b",
        'Colony': "//td/h2[contains(text(), 'COLONIA')]/span/b",
        'Price': "//td[@align='right']/h2/b",
        'Publication Date': "//td[@class='carac_td' and contains(text(), 'Publicado')]",
        'Bedrooms': "//td[@class='carac_td']/h3[contains(text(), 'Recámaras')]",
        'Bathrooms': "//td[@class='carac_td']/h3[contains(text(), 'Baños')]",
        'Construction Size': "//td[@class='carac_td']/h3[contains(text(), 'Construcción')]",
        'Land Size': "//td[@class='carac_td']/h3[contains(text(), 'Terreno')]",
        'Floors': "//td[@class='carac_td']/h3[contains(text(), 'Planta')]",
        'Phone': "//td[@class='carac_td' and img[@alt='Telefono']]"
    }

    for key, xpath in xpaths.items():
        try:
            element_text = WebDriverWait(driver, timeout).until(
                EC.presence_of_element_located((By.XPATH, xpath))
            ).text.strip()

            # Convert specific fields to integers
            if key in ['Bedrooms', 'Bathrooms', 'Construction Size', 'Land Size', 'Floors']:
                element_text = int(re.sub(r'[^\d]', '', element_text))

            # Clean price text
            if key == 'Price':
                element_text = int(re.sub(r'[^\d]', '', element_text))

            # Convert Publication Date to datetime
            if key == 'Publication Date':
                element_text = convert_to_date(element_text)

            house_details[key] = element_text
        except TimeoutException:
            house_details[key] = None

    house_details['URL'] = link
    houses_data.append(house_details)

# Close the browser
driver.quit()
```

# Creating DF to analyze the data

```python
# Close the browser
driver.quit()

# Create a DataFrame from the collected data
df = pd.DataFrame(houses_data)
```

```python
# Create a DataFrame from the collected data
df = pd.DataFrame(houses_data)

# Save the DataFrame to a CSV file
df.to_csv('houses_data_elite.csv', index=False)
```

```python
# Calculate the value per square meter for each house
df['Value per m²'] = df.apply(lambda row: row['Price'] / (row['Construction Size'] + row['Land Size']), axis=1)

# Calculate the overall average market value (AMC) per square meter
overall_amc = df['Value per m²'].mean()

# Calculate the difference from AMC
df['Difference from AMC'] = abs(df['Value per m²'] - overall_amc)

# Convert 'Publication Date' to datetime
df['Publication Date'] = pd.to_datetime(df['Publication Date'])

# Calculate the recency score (days since publication)
current_date = datetime.now()
df['Days Published'] = (current_date - df['Publication Date']).dt.days

# Calculate recency factor (more recent = higher likelihood)
df['Recency Factor'] = 1 / (1 + df['Days Published'])

# Normalize the difference from AMC to get the likelihood factor (closer to AMC = higher likelihood)
max_difference = df['Difference from AMC'].max()
df['Likelihood Factor'] = 1 - (df['Difference from AMC'] / max_difference)

# Combine recency factor and likelihood factor to estimate the likelihood of being sold
df['Likelihood of Sale'] = df['Recency Factor'] * df['Likelihood Factor']

# Save the updated DataFrame with the likelihood of sale to a CSV file
df.to_csv('houses_data_elite_amc.csv', index=False)

# Print the overall AMC for reference
print(f'Overall AMC per square meter: {overall_amc:.2f}')
```

```python
# Add a new column for the value per square meter
df['Value per m²'] = df.apply(lambda row: row['Price'] / (row['Construction Size'] + row['Land Size']), axis=1)

# Calculate the average market value (AMC) per square meter
average_amc = df['Value per m²'].mean()

# Add a new column for the publication date (converted to datetime)
df['Publication Date'] = pd.to_datetime(df['Publication Date'])

# Calculate the duration since publication in days
df['Days Published'] = (datetime.now() - df['Publication Date']).dt.days

# Normalize 'Days Published' to a scale (e.g., 0 to 1)
max_days_published = df['Days Published'].max()
df['Normalized Days Published'] = df['Days Published'] / max_days_published

# Calculate the likelihood of being sold
# The closer the 'Value per m²' is to the 'average_amc', the higher the likelihood
# The longer it has been published, the lower the likelihood

def likelihood(row):
    distance_from_amc = abs(row['Value per m²'] - average_amc)
    likelihood_score = max(0, 1 - distance_from_amc / average_amc) * (1 - row['Normalized Days Published'])
    return likelihood_score

df['Likelihood of Being Sold'] = df.apply(likelihood, axis=1)

# Save the DataFrame to a CSV file
df.to_csv('houses_data_likely_hood.csv', index=False)
```
